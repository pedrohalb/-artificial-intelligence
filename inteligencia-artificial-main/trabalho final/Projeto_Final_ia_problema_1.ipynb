## Treinamento/validação dos modelos

Nesta seção, treinamos e validamos diferentes modelos de aprendizado de máquina para avaliar sua performance em nosso conjunto de dados. Utilizamos três modelos distintos: K-Nearest Neighbors (KNN), Decision Tree (Árvore de Decisão) e Random Forest.

Para cada modelo, definimos uma grade de hiperparâmetros a serem testados e utilizamos validação cruzada para avaliar a performance de cada combinação de hiperparâmetros. A função `train_models` realiza o treinamento e a avaliação dos modelos, repetindo o processo várias vezes para obter médias mais confiáveis das métricas de desempenho. As métricas avaliadas incluem acurácia, recall, precisão, F1 Score e matriz de confusão.

A função `print_results_train` é responsável por imprimir os resultados de cada modelo, mostrando as médias das métricas de desempenho para cada combinação de hiperparâmetros testada.


```python
from sklearn.model_selection import ParameterGrid
import numpy as np
```

```python
from sklearn.model_selection import train_test_split, ParameterGrid
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
import numpy as np

def train_models(model_class, param_grid, train_data, n):
    dadosEntrada_train, saidaDesejada_train = train_data
    results_model1 = []
    results_model2 = []

    # Gerando combinações de hiperparâmetros
    for params in ParameterGrid(param_grid):
        models1 = []
        accuracy_scores_model1 = []
        recall_scores_model1 = []
        precision_scores_model1 = []
        f1_scores_model1 = []
        confusion_matrices_model1 = []

        models2 = []
        accuracy_scores_model2 = []
        recall_scores_model2 = []
        precision_scores_model2 = []
        f1_scores_model2 = []
        confusion_matrices_model2 = []

        # Repetindo a execução n vezes
        for _ in range(n):
            # Dividindo o conjunto de treinamento em dois subconjuntos disjuntos
            dadosEntrada_train1, dadosEntrada_train2, saidaDesejada_train1, saidaDesejada_train2 = train_test_split(
                dadosEntrada_train, saidaDesejada_train, test_size=0.5)

            # Treinando o primeiro modelo
            model1 = model_class(**params)
            model1.fit(dadosEntrada_train1, saidaDesejada_train1)  # Treinando com um subconjunto
            saidaPrevista1 = model1.predict(dadosEntrada_train2)  # Testando com o outro subconjunto

            models1.append(model1)
            accuracy_scores_model1.append(accuracy_score(saidaDesejada_train2, saidaPrevista1))
            recall_scores_model1.append(recall_score(saidaDesejada_train2, saidaPrevista1))
            precision_scores_model1.append(precision_score(saidaDesejada_train2, saidaPrevista1))
            f1_scores_model1.append(f1_score(saidaDesejada_train2, saidaPrevista1))
            confusion_matrices_model1.append(confusion_matrix(saidaDesejada_train2, saidaPrevista1))

            # Treinando o segundo modelo
            model2 = model_class(**params)
            model2.fit(dadosEntrada_train2, saidaDesejada_train2)  # Treinando com o outro subconjunto
            saidaPrevista2 = model2.predict(dadosEntrada_train1)  # Testando com o primeiro subconjunto

            models2.append(model2)
            accuracy_scores_model2.append(accuracy_score(saidaDesejada_train1, saidaPrevista2))
            recall_scores_model2.append(recall_score(saidaDesejada_train1, saidaPrevista2))
            precision_scores_model2.append(precision_score(saidaDesejada_train1, saidaPrevista2))
            f1_scores_model2.append(f1_score(saidaDesejada_train1, saidaPrevista2))
            confusion_matrices_model2.append(confusion_matrix(saidaDesejada_train1, saidaPrevista2))

        # Calculando a média das métricas para os dois modelos
        results_model1.append({
            'models': models1,
            'params': params,
            'accuracy': np.mean(accuracy_scores_model1),
            'recall': np.mean(recall_scores_model1),
            'precision': np.mean(precision_scores_model1),
            'f1': np.mean(f1_scores_model1),
            'confusion_matrix': np.mean(confusion_matrices_model1, axis=0)
        })

        results_model2.append({
            'models': models2,
            'params': params,
            'accuracy': np.mean(accuracy_scores_model2),
            'recall': np.mean(recall_scores_model2),
            'precision': np.mean(precision_scores_model2),
            'f1': np.mean(f1_scores_model2),
            'confusion_matrix': np.mean(confusion_matrices_model2, axis=0)
        })

    return results_model1, results_model2
```

```python
# Função para imprimir os resultados
def print_results_train(results, model_name):
    print(f"\nResultados para o modelo {model_name}:")
    for result in results:
        print(f"Parâmetros: {result['params']}")
        print(f"Acurácia média: {result['accuracy']:.4f}")
        print(f"Recall médio: {result['recall']:.4f}")
        print(f"Precisão média: {result['precision']:.4f}")
        print(f"F1 Score médio: {result['f1']:.4f}")
        print("-" * 30)
        print("--- Matriz de Confusão ---")
        print(pd.DataFrame(result['confusion_matrix'], index=['neg', 'pos'], columns=['pred_neg', 'pred_pos']))
        print("-" * 30)
```

### KNN

Primeiro, definimos os hiperparâmetros e avaliamos o modelo K-Nearest Neighbors:

- `knn_param_grid`: Contém diferentes valores para o número de vizinhos (`n_neighbors`) e a métrica de distância (`metric`) a serem testados.
- `train_models`: Treina e avalia o modelo KNN para cada combinação de hiperparâmetros, executando o processo apenas uma vez para cada combinação.
- `print_results_train`: Imprime os resultados da avaliação para cada combinação de hiperparâmetros.


```python
# Definindo os hiperparâmetros para o knn
knn_param_grid = {'n_neighbors': [3, 5, 7, 9, 11], 'metric': ['euclidean', 'manhattan', 'chebyshev']}


# Avaliando knn nos dois subconjuntos de treinamento
knn_results1_train, knn_results2_train  = train_models(KNeighborsClassifier, knn_param_grid, (dadosEntrada_train, saidaDesejada_train), 1)

# Resultados
print_results_train(knn_results1_train, "K-Nearest Neighbors (Subconjunto 1)")
print_results_train(knn_results2_train, "K-Nearest Neighbors (Subconjunto 2)")
```

### Árvore de Decisão


Em seguida, repetimos o mesmo processo para o modelo de Árvore de Decisão:

- `dt_param_grid`: Contém diferentes valores para a profundidade máxima da árvore (`max_depth`) e o número mínimo de amostras necessárias para dividir um nó (`min_samples_split`).
- `train_models`: Treina e avalia o modelo de Árvore de Decisão para cada combinação de hiperparâmetros, repetindo o processo 100 vezes para cada combinação.
- `print_results_train`: Imprime os resultados da avaliação para cada combinação de hiperparâmetros.


```python
from sklearn.tree import DecisionTreeClassifier

# Definindo os hiperparâmetros para a Árvore de Decisão
dt_param_grid = {'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10, 15, 20]}

# Avaliando a Árvore de Decisão
dt_results1, dt_results2 = train_models(DecisionTreeClassifier, dt_param_grid, (dadosEntrada_train, saidaDesejada_train), 100)

# # Imprimindo os resultados para a Árvore de Decisão
print_results_train(dt_results1, "Árvore de Decisão (Subconjunto 1)")
print_results_train(dt_results2, "Árvore de Decisão (Subconjunto 2)")
```

### Random Forest

Por fim, realizamos o treinamento e avaliação do modelo Random Forest:

- `rf_param_grid`: Contém diferentes valores para o número de estimadores (`n_estimators`) e a profundidade máxima das árvores (`max_depth`).
- `train_models`: Treina e avalia o modelo Random Forest para cada combinação de hiperparâmetros, repetindo o processo 100 vezes para cada combinação.

```python
# Definindo os hiperparâmetros o random forest
rf_param_grid = {'n_estimators': [50, 100, 150, 200, 250], 'max_depth': [None, 10, 20]}

# Avaliando random forest
rf_results1, rf_results2 = train_models(RandomForestClassifier, rf_param_grid, (dadosEntrada_train, saidaDesejada_train), 100)

# resultados
print_results_train(rf_results1, "Random Forest (Subconjunto 1)")
print_results_train(rf_results2, "Random Forest (Subconjunto 2)")
```

Para cada modelo, a função `test_models` realiza a avaliação utilizando os modelos treinados e calcula as métricas de desempenho, como acurácia, recall, precisão, F1 Score e matriz de confusão. Em seguida, a função `print_results_test` é responsável por imprimir os resultados obtidos.


```python
from sklearn.model_selection import ParameterGrid
import numpy as np
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
import numpy as np

def test_models(models, test_data):
    dadosEntrada_test, saidaDesejada_test = test_data
    results = []

    for model_info in models:
        params = model_info['params']
        models_trained = model_info['models']  # Obtendo os modelos já treinados

        accuracy_scores = []
        recall_scores = []
        precision_scores = []
        f1_scores = []
        confusion_matrices = []

        # Avaliando cada modelo treinado no conjunto de teste
        for model in models_trained:
            saidaPrevista = model.predict(dadosEntrada_test)

            # Calculando as métricas de avaliação
            accuracy_scores.append(accuracy_score(saidaDesejada_test, saidaPrevista))
            recall_scores.append(recall_score(saidaDesejada_test, saidaPrevista))
            precision_scores.append(precision_score(saidaDesejada_test, saidaPrevista))
            f1_scores.append(f1_score(saidaDesejada_test, saidaPrevista))
            confusion_matrices.append(confusion_matrix(saidaDesejada_test, saidaPrevista))

        # Calculando a média das métricas para os modelos treinados
        results.append({
            'params': params,
            'accuracy': np.mean(accuracy_scores),
            'recall': np.mean(recall_scores),
            'precision': np.mean(precision_scores),
            'f1': np.mean(f1_scores),
            'confusion_matrix': np.mean(confusion_matrices, axis=0)
        })

    return results
```

```python
def print_results_test(results, model_name):
    print(f"\nResultados para o modelo {model_name} (Teste):")
    for result in results:
        print(f"Parâmetros: {result['params']}")
        print(f"Acurácia: {result['accuracy']:.4f}")
        print(f"Recall: {result['recall']:.4f}")
        print(f"Precisão: {result['precision']:.4f}")
        print(f"F1 Score: {result['f1']:.4f}")
        print("-" * 30)
        print("--- Matriz de Confusão ---")
        print(pd.DataFrame(result['confusion_matrix'], index=['neg', 'pos'], columns=['pred_neg', 'pred_pos']))
        print("-" * 30)
```

### Regra de decisão (Meu primeiro Classificador)

Além dos modelos tradicionais, também testamos um modelo baseado em regras simples, que utiliza um limiar para classificação a partir do cálculo do máximo valor de cada linha do conjunto:

- `test_rule_based_model`: Avalia o modelo baseado em regras utilizando o conjunto de teste e calcula as métricas de desempenho.
- `print_evaluation_results`: Imprime os resultados da avaliação para o modelo baseado em regras.


```python
def simple_rule_based_model(row, threshold):
    return 'B' if row.max() < threshold else 'M'
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix

def test_rule_based_model(test_data, threshold):
    dadosEntrada_test, saidaDesejada_test = test_data

    saidaPrevista_test = dadosEntrada_test.apply(lambda row: simple_rule_based_model(row, threshold), axis=1)
    saidaPrevista_test = saidaPrevista_test.map({'B': 0, 'M': 1})

    accuracy = accuracy_score(saidaDesejada_test, saidaPrevista_test)
    recall = recall_score(saidaDesejada_test, saidaPrevista_test)
    precision = precision_score(saidaDesejada_test, saidaPrevista_test)
    f1 = f1_score(saidaDesejada_test, saidaPrevista_test)
    confusion_matrix_result = confusion_matrix(saidaDesejada_test, saidaPrevista_test)

    results = {
        'accuracy': accuracy,
        'recall': recall,
        'precision': precision,
        'f1': f1,
        'confusion_matrix': confusion_matrix_result
    }

    return results
```

```python
import pandas as pd

def print_evaluation_results(results, model_name):
    print(f"\nResultados para o modelo {model_name}:")
    print(f"Acurácia: {results['accuracy']:.4f}")
    print(f"Recall: {results['recall']:.4f}")
    print(f"Precisão: {results['precision']:.4f}")
    print(f"F1 Score: {results['f1']:.4f}")
    print("--- Matriz de Confusão ---")
    print(pd.DataFrame(results['confusion_matrix'], index=['neg', 'pos'], columns=['pred_neg', 'pred_pos']))
```

####Comparação do Recall

Agora, nesta seção iremos comparar o recall de cada modelo em relação ao problema, pelo fato de que um bom resultado no treinamento não necessariamente significa que o seu modelo terá um bom resultado no teste.

No caso da detecção do câncer de mama o objetivo principal é maximizar o Recall que por sua vez significa maximizar o número de verdadeiros positivos e minimizar o número de falsos negativos.

Para isso utilizaremos a função `plot_recall_bar` que cria um gráfico de barras que compara os valores de recall para diferentes configurações de hiperparâmetros de um modelo, permitindo uma visualização clara de como os hiperparâmetros afetam o desempenho do modelo em termos de recall. Como também a função `plot_recall_line` que cria um gráfico de linha que mostra os valores de recall de diferentes modelos ou configurações de hiperparâmetros, permitindo uma visualização clara de como esses valores variam entre os modelos. O gráfico é personalizado com rótulos, título e legenda, e os pontos de dados são marcados e conectados por uma linha. E por fim, a função `compare_max_min_recall` que compara os valores de recall máximo e mínimo para diferentes configurações de hiperparâmetros de um modelo, plota esses valores em um gráfico de barras, adiciona rótulos para os valores e hiperparâmetros correspondentes, e retorna a razão entre o recall máximo e mínimo. Isso permite visualizar e entender a variação no desempenho do modelo com diferentes configurações de hiperparâmetros.







```python
import matplotlib.pyplot as plt
import numpy as np


def plot_recall_bar(results, model_name):
    params_list = [str(res['params']) for res in results]
    recalls = [res['recall'] for res in results]
    num_params = len(results)

    x = np.arange(num_params)  # the label locations
    width = 0.5  # the width of the bars

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(x, recalls, width, label='Recall')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_xlabel('Hyperparameters')
    ax.set_ylabel('Recall Score')
    ax.set_title(f'Recall Comparison for {model_name}')
    ax.set_xticks(x)
    ax.set_xticklabels(params_list, rotation=45, ha="right")
    ax.legend()

    fig.tight_layout()
    plt.show()
```

```python
def plot_recall_line(results, model_name):
    recalls = [res['recall'] for res in results]
    model_numbers = [f'M {i+1}' for i in range(len(results))]

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(model_numbers, recalls, marker='o', linestyle='-', label='Recall')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_xlabel('Model')
    ax.set_ylabel('Recall Score')
    ax.set_title(f'Recall Comparison for {model_name}')
    ax.legend()

    fig.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

def compare_max_min_recall(results, model_name):
    recalls = [res['recall'] for res in results]
    max_recall = max(recalls)
    min_recall = min(recalls)
    max_recall_params = [res['params'] for res in results if res['recall'] == max_recall][0]
    min_recall_params = [res['params'] for res in results if res['recall'] == min_recall][0]
    ratio = max_recall / min_recall

    # Plotting
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.bar(['Max Recall', 'Min Recall'], [max_recall, min_recall], color=['blue', 'red'])

    # Adding text labels for the values
    ax.text(0, max_recall + 0.01, f'{max_recall:.2f}\n{max_recall_params}', ha='center', va='bottom', fontsize=10)
    ax.text(1, min_recall + 0.01, f'{min_recall:.2f}\n{min_recall_params}', ha='center', va='bottom', fontsize=10)

    ax.set_xlabel('Recall Type')
    ax.set_ylabel('Recall Score')
    ax.set_title(f'Comparison of Max and Min Recall for {model_name}')
    ax.set_ylim(0, 1)  # Assuming recall is between 0 and 1

    plt.tight_layout()
    plt.show()

    return ratio
```

### Geral

Essa seção apresenta uma análise comparativa dos resultados gerais de desempenho e recall entre os diferentes modelos de aprendizado de máquina. Os gráficos gerados fornecem uma visão geral de como os modelos se comparam em termos de métricas de avaliação, tanto nos conjuntos de treinamento quanto nos conjuntos de teste.



As chamadas da função `plot_general_comparison` fornecem uma análise dos valores máximos e mínimos de recall para diferentes conjuntos de resultados de treinamento e teste de cada modelo. Isso ajuda a entender como os modelos se comportam em termos de recall em diferentes conjuntos de dados e como a variação dos hiperparâmetros pode afetar a consistência do desempenho do modelo.

```python
def plot_general_comparison(train_results, test_results, model_names, metric='recall'):
    train_max_scores = []
    train_min_scores = []
    train_max_params = []
    train_min_params = []

    test_max_scores = []
    test_min_scores = []
    test_max_params = []
    test_min_params = []

    for train, test in zip(train_results, test_results):
        train_metrics = [res[metric] for res in train]
        train_max_score_idx = np.argmax(train_metrics)
        train_min_score_idx = np.argmin(train_metrics)
        train_max_scores.append(train_metrics[train_max_score_idx])
        train_min_scores.append(train_metrics[train_min_score_idx])
        train_max_params.append(train[train_max_score_idx]['params'])
        train_min_params.append(train[train_min_score_idx]['params'])

        test_metrics = [res[metric] for res in test]
        test_max_score_idx = np.argmax(test_metrics)
        test_min_score_idx = np.argmin(test_metrics)
        test_max_scores.append(test_metrics[test_max_score_idx])
        test_min_scores.append(test_metrics[test_min_score_idx])
        test_max_params.append(test[test_max_score_idx]['params'])
        test_min_params.append(test[test_min_score_idx]['params'])

    num_models = len(model_names)
    x = np.arange(num_models)
    width = 0.35

    fig, ax = plt.subplots(2, figsize=(12, 12))

    rects1 = ax[0].bar(x - width/2, train_max_scores, width, label='Best Train', color='b')
    rects2 = ax[0].bar(x + width/2, train_min_scores, width, label='Worst Train', color='r')

    rects3 = ax[1].bar(x - width/2, test_max_scores, width, label='Best Test', color='g')
    rects4 = ax[1].bar(x + width/2, test_min_scores, width, label='Worst Test', color='y')

    ax[0].set_xlabel('Models')
    ax[0].set_ylabel(metric.capitalize())
    ax[0].set_title(f'Training Comparison of {metric.capitalize()} among Models')
    ax[0].set_xticks(x)
    ax[0].set_xticklabels(model_names, rotation=45, ha="right")
    ax[0].legend()

    ax[1].set_xlabel('Models')
    ax[1].set_ylabel(metric.capitalize())
    ax[1].set_title(f'Testing Comparison of {metric.capitalize()} among Models')
    ax[1].set_xticks(x)
    ax[1].set_xticklabels(model_names, rotation=45, ha="right")
    ax[1].legend()

    def autolabel(rects, scores, params, color, ax):
        for i, rect in enumerate(rects):
            height = rect.get_height()
            ax.annotate(f'{scores[i]:.2f}\n{params[i]}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', color=color)

    autolabel(rects1, train_max_scores, train_max_params, 'b', ax[0])
    autolabel(rects2, train_min_scores, train_min_params, 'r', ax[0])
    autolabel(rects3, test_max_scores, test_max_params, 'g', ax[1])
    autolabel(rects4, test_min_scores, test_min_params, 'y', ax[1])

    fig.tight_layout()
    plt.show()
```

```python
# Função para plotar a comparação geral de métricas entre os modelos
plot_general_comparison([knn_results1_train, dt_results1, rf_results1],
                        [knn_results_test1, dt_results_test1, rf_results_test1],
                        ["Logistic Regression", "Decision Tree", "Random Forest"])
```

### Conclusão

A importância de utilizar diferentes modelos em problemas de aprendizado supervisionado não pode ser subestimada. Cada modelo tem suas próprias forças e fraquezas, e aplicar uma variedade de algoritmos pode fornecer uma visão mais completa e precisa do problema em questão. Modelos como K-NN, Árvore de Decisão e Random Forest têm se mostrado eficazes em diferentes contextos de detecção do câncer de mama, contribuindo para diagnósticos mais rápidos e precisos.

O impacto de diferentes conjuntos de treinamento sobre o desempenho do modelo é significativo. Dados de alta qualidade e representativos são essenciais para treinar modelos robustos e precisos. Mudanças nos conjuntos de treinamento podem influenciar drasticamente a capacidade do modelo de generalizar e fazer previsões corretas em dados novos.

Além disso, a escolha dos hiperparâmetros é crucial para o desempenho do modelo. Hiperparâmetros mal ajustados podem levar a modelos que sofrem de overfitting ou underfitting, resultando em baixa precisão. Ajustes cuidadosos e testes iterativos de diferentes configurações de hiperparâmetros são necessários para otimizar o desempenho dos modelos de aprendizado de máquina.

Porém é importante lembrar que um bom resultado no treinamento não necessariamente significa que o seu modelo terá um bom resultado no teste. No caso da detecção do câncer de mama o objetivo principal é maximizar o Recall que por sua vez significa maximizar o número de verdadeiros positivos e minimizar o número de falsos negativos.

Podemos observar nos gráfios de comparação do recall que o modelo KNN demonstrou valores maiores que os outros modelos (0.94 e 0.89) no teste, o que indica que ele é adequado para problemas como o do câncer de mama que necessitam de um valor alto em relação ao recall.

Em resumo, a aplicação de diferentes algoritmos de aprendizado de máquina na detecção do câncer de mama pode salvar vidas ao aumentar a precisão e a velocidade dos diagnósticos. A seleção adequada de conjuntos de dados e o ajuste dos hiperparâmetros são etapas cruciais nesse processo, garantindo que os modelos desenvolvidos sejam eficazes e confiáveis.
