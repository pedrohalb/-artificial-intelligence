## Seleção e preparação dos dados;

A saída esperada na coluna `diagnosis` é categórica, isto é *M* representa os tumores malignos e *B* os benignos. Para apliar o modelo K-NN nessa base é necessário codificar as os valores *M* e *B* em variáveis categoricas numéricas. Existem várias formas de fazer esse processo em Python, uma delas é usando o `LabelEncoder`.

Nesta etapa de preparação dos dados é necessário separar os dados brutos visando criar subconjuntos para treino, validação e teste do modelo. Existem algumas formas de executar esse procedimento. Uma delas que é simples e eficaz é usar função o `train_test_split`, do pacote `Scikit-learn`.

```
# Separando os dados em entrada e saida desejada e removendo as colunas `diagnosis` e `id`
dadosEntrada = df.drop(['diagnosis', 'id'], axis=1)
saidaDesejada = df['diagnosis']

# Aplicando a função LabelEncoder na variavel saidaDesejada
saidaDesejada = LabelEncoder().fit_transform(saidaDesejada)

# Aplicando a funcao train_test_split para dividir o conjunto original em 70% para treindo e 30% para teste
dadosEntrada_train, dadosEntrada_test, saidaDesejada_train, saidaDesejada_test = train_test_split(dadosEntrada, saidaDesejada, test_size=0.3)
```

Não é o caso específico do K-NN mas em alguns modelos (Redes Neuras Artificiais, por exemplo)  é necessário realizar transformações os dados/variáveis visando mapeá-los e um conjunto que possui um dominio específico. Esse procedimento é conhecido como *normalização dos dados* e visa a maximização do aprendizado do modelo.



## Treinamento/validação do modelo

Como descrito anteriormente, o `K-NN` foi o modelo selecionado para solucionar o problema de detecção do cancer de mama na base de dados selecionada.

Na literatura existem diversos modelos que poderiam ser utilizados para a mesma finalidade. O K-NN foi escolhido devido a sua facilidade de uso e entendimento.

Vale ressaltar que, mesmo sendo um modelo com baixo grau de complexidade, o `K-NN` apresenta resultados interessantes para grande parte dos problemas no qual ele é aplicado.

O `K-NN` será executado na versão implementada na biblioteca `Scikit-learn` através da função `KNeighborsClassifier`. Para ter acesso a todos os hiperparâmetros do modelo K-NN consulte o [manual do modelo](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html) no portal do `Scikit-learn`.

```
# Criando uma instancia do K-NN
# Nessa simulacao o número de vizinhos n_neighbors=10 foi escolhido empiricamente a métrica de distância escolhida foi a distancia euclidiana
knnModel = KNeighborsClassifier(n_neighbors=4, metric='euclidean')

# Ajustando o modelo K-NN com os dados de treino
knnModel.fit(dadosEntrada_train, saidaDesejada_train)

# Calculando a saida para o arquivo de treinamento (classificando os dados de treino)
saidaPrevista_train = knnModel.predict(dadosEntrada_train)

# Verificando o erro em relacao ao arquivo de treinamento
accuracy_train = accuracy_score(saidaDesejada_train, saidaPrevista_train)
recall_train = recall_score(saidaDesejada_train, saidaPrevista_train)
precision_train = precision_score(saidaDesejada_train, saidaPrevista_train)
f1_train = f1_score(saidaDesejada_train, saidaPrevista_train)
print('------------------------------')
print('------------------------------')
print('Acurácia: ', accuracy_train)
print('Recall: ', recall_train)
print('Precision: ', precision_train)
print('f1 Score: ', f1_train)
print('------------------------------')
print('------------------------------')
print('--- Matriz de Confusão ---')
pd.DataFrame(confusion_matrix(saidaDesejada_train, saidaPrevista_train),
             index=['neg', 'pos'], columns=['pred_neg', 'pred_pos'])
```

## Verificação do desempenho

Um bom resultado no treinamento não necessáriamente significa que o seu modelo terá um bom resultado no teste.

No caso da detecção do câncer de mama o objetivo principal é maximizar o Recall que por sua vez significa maximizar o número de verdadeiros positivos e minimizar o número de falsos negativos. Encontrar um ajuste ideal nem sempre é uma tarefa trival e pode exigir muitas horas de dedicação dependendo problema a ser resolvido.

Aqui, além da métrica de acurácia serão analisadas as métricas de precisão, *recall*, *f1-score* e a  *matriz de confusão*.

```
# Aplicando o modelo ajustado aos dados de teste
saidaPrevista_test = knnModel.predict(dadosEntrada_test)

# Verificando o erro em relação ao conjunto de testes
accuracy_test = accuracy_score(saidaDesejada_test, saidaPrevista_test)
recall_test = recall_score(saidaDesejada_test, saidaPrevista_test)
precision_test = precision_score(saidaDesejada_test, saidaPrevista_test)
f1_test = f1_score(saidaDesejada_test, saidaPrevista_test)
print('------------------------------')
print('------------------------------')
print('Acurácia: ', accuracy_test)
print('Recall: ', recall_test)
print('Precision: ', precision_test)
print('f1 Score: ', f1_test)
print('------------------------------')
print('------------------------------')
print('--- Matriz de Confusão ---')
pd.DataFrame(confusion_matrix(saidaDesejada_test, saidaPrevista_test),
             index=['neg', 'pos'], columns=['pred_neg', 'pred_pos'])
```

## Considerações


Se observarmos as métricas do modelo criado com o algoritmo K-NN é possível inferir que valores foram promissores em relação as métricas consideradas.

Por outro lado, a matriz de confusão mostra que no conjunto de teste o modelo cometeu um erros graves de diagnóstico com falsos negativos o que é crítico em problemas que envolvem saúde pública.

Em linhas gerais, é primordial trabalhar com a matriz de confusão visando minimizar os falsos negativos aumentando, se necessário, as classificações com falso positivo. Lembrando que essa afirmação pode variar para cada tipo de problema em análise.
